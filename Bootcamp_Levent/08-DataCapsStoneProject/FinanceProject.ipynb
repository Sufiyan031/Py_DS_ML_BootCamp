```python
from pandas_datareader import data, wb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
import seaborn as sns


sns.set()
%matplotlib inline
```

```python
df = pd.read_pickle('all_banks')
df.head()
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
df.xs('Close',axis=1,level=1).max()
#df.xs('Close',axis=1,level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+' Return'] = df[tick]['Close'].pct_change()

returns.head()
```

```python
sns.pairplot(data=returns[1:])
```

```python
returns.idxmin()
```

```python
returns.idxmax()
```

```python
returns.std()
```

```python
returnsRI = returns.reset_index()
returns2015 = returnsRI[returnsRI['Date'].dt.year==2015]
returns2015.std()
```

```python
sns.distplot(returns2015['MS Return'],bins=50)
```

```python
returns2008 = returnsRI[returnsRI['Date'].dt.year == 2008]
sns.distplot(returns2008['C Return'],bins=100)
```

```python
for tick in tickers:         
    df[tick]['Close'].plot(figsize=(12,4),label=tick)

plt.legend()    
```
