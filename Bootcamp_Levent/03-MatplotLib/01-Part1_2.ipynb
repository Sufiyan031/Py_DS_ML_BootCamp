```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#%matplotlib inline #matplot lib grafiklerini jupyter notebook icinde gormemize yariyor
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
#FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#ObjectOriented Method

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,1,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Small Plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
'''for current in axes:
    current.plot(x,y)
'''
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Pilot')

axes[1].plot(y,x)
axes[1].set_title('Second Pilot')
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))


axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_pic.png',dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x square')
ax.plot(x,x**3,label='x cube')

ax.legend()
```
