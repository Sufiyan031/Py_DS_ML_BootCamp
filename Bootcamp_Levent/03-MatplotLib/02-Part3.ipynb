```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth = 5,alpha=0.8,linestyle='--',marker='o',markersize=15
        ,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
#ax.plot(x,y,color='blue',lw = 5,alpha=0.1,ls='-.')
#ax.plot(x,y,color='#FF8C00') #RGB Hex Code
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw = 2,ls='--')
ax.set_xlim([0,1,])
ax.set_ylim([0,2])
```
